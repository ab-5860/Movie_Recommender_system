```python
#Literal_eval will convert string into list and No exception will be thrown
import ast
def convert(obj):
  L =[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  
  return L
```

```python
movie['genres'] = movie['genres'].apply(convert)
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
```

```python
import ast
def convert4(obj):
  L =[]
  count=0
  for i in ast.literal_eval(obj):
    if count!=4:
      L.append(i['name'])
      count+=1
    else:
      break
  
  return L
```

```python
movie['cast'] = movie['cast'].apply(convert4)
```

```python
def convert_dir(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  
  return L
```

```python
movie['crew'] = movie['crew'].apply(convert_dir)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')   #CountVectorizer converts into scipy sparse matrix
```

```python
cv.fit_transform(new_df1['tags']).toarray().shape
```

```python
vector = cv.fit_transform(new_df1['tags']).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
cv.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  L =[]
  for i in text.split():
    L.append(ps.stem(i))

  return " ".join(L)

```

```python
new_df1['tags'] = new_df1['tags'].apply(stem)
```

```python
#Similarity between vectors

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
cosine_similarity(vector).shape
```

```python
similarity[0]
```

```python
new_df1[new_df1['title']=='Batman Begins'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df1[new_df1['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df1.iloc[i[0]].title)
```

```python
recommend('Batman')
```
